```python
def format_zero_shot_prompt(question, options):
    """
    Format the prompt for Zero-Shot prompting with explicit instruction to return only the option letter.
    """
    prompt = (
        f"Choose the correct Option letter to the given question from the options below and complete the following. "
        f"Question: {question}\n"
        f"A) {options[0]}\n"
        f"B) {options[1]}\n"
        f"C) {options[2]}\n"
        f"D) {options[3]}\n"
        f"correct option is = ("
    )
    return prompt

def format_chain_of_thought_prompt(question, options):
    """
    Format the prompt for Chain of Thought (Zero-Shot) prompting with instruction to return only the option letter.
    """
    prompt = (
        f"Choose the correct Option letter to the given question from the options below. "
        f"Think step by step, but at the end, respond only with the letter corresponding to the correct option.\n\n"
        f"Question: {question}\n"
        f"A) {options[0]}\n"
        f"B) {options[1]}\n"
        f"C) {options[2]}\n"
        f"D) {options[3]}\n"
        f"correct option is = ("
    )
    return prompt
```

```python
import torch

def infer_result(prompt, tokenizer, model):
    """
    Generate model output for the given prompt.
    """
    inputs = tokenizer(prompt, return_tensors="pt")
    
    # Move inputs to the correct device
    inputs = {key: val.to(model.device) for key, val in inputs.items()}
    
    # Generate output (adjust max_length as needed)
    with torch.no_grad():
        output = model.generate(**inputs, max_new_tokens=75)
        
    # Decode the output tokens
    result = tokenizer.decode(output[0], skip_special_tokens=True)
    
#     print("result is", result)
    
    # Post-process to return only the letter inside parentheses
    start = result.find("correct option is = (")
    end = result.find(")", start)
    
    if start != -1 and end != -1:
        return result[start+21:end].strip()  # Return only the content inside the parentheses
    return result.strip()  # Return the entire output if parentheses not found
```

```python
import time

def evaluate(model_answer, correct_answer):
    """
    Compare the model's answer with the correct one.
    """
    return model_answer.strip() == correct_answer.strip()

def run_evaluation(example, prompt_function, tokenizer, model):
    """
    Run the evaluation for a given example using the provided model and tokenizer.
    """
    question = example['question']
    options = [example['choices'][0], example['choices'][1], example['choices'][2], example['choices'][3]]
    correct_answer = example['answer']
    dict_options = {0:'A', 1:'B', 2:'C', 3:'D'}
    correct_answer = dict_options[correct_answer]
    
    # Format the prompt
    prompt = prompt_function(question, options)
     
    start_time = time.time()
    # Get the model's result
    model_answer = infer_result(prompt, tokenizer, model)
    end_time = time.time()
    
    # Calculate inference time
    inference_time = end_time - start_time
    
    # Evaluate whether the answer is correct
    is_correct = evaluate(model_answer, correct_answer)
    return is_correct, inference_time
```

```python
from datasets import load_dataset

ds = load_dataset("cais/mmlu", "abstract_algebra")
```

```python
def evaluate_dataset(dataset_split, prompt_function, tokenizer, model):
    """
    Iterate over the dataset, evaluate each example, count correct answers, and time each iteration.
    """
    correct_count = 0
    total_time = 0
    num_examples = len(dataset_split)
    counter = 1

    for example in dataset_split:
        print(counter, end='\r')
        # Extract the question and options
        question = example['question']
        options = example['choices']
        correct_answer = example['answer']

        # Run the evaluation
        is_correct, inference_time = run_evaluation(example, prompt_function, tokenizer, model)

        # Update correct count and total time
        if is_correct:
            correct_count += 1
        total_time += inference_time
        
        counter+=1

    # Print final results
    print(f"Total Correct Answers: {correct_count}/{num_examples}")
    print(f"Average Inference Time per Example: {total_time / num_examples:.4f} seconds")
    
    
    return (total_time / num_examples), (correct_count/num_examples)
```

```python
avg_inference_time_zero_gemma, correct_frac_zero_gemma = evaluate_dataset(ds['test'], format_zero_shot_prompt, tokenizer_gemma, model_gemma)
```
